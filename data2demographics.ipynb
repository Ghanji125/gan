```python
import numpy as np
import pandas as pd
from collections import Counter

df = pd.read_excel("mydata4.xlsx")
df["stage_at_diagnosis"] = df["stage_at_diagnosis"].replace(to_replace=["IIA", "IIB"], value="II")
new_df = df.fillna("unknown")
new_df.rename(columns={'Unnamed: 0':'id'}, inplace = True)
colname = new_df.columns
for k,v in enumerate(colname):
    print(k, ": ", v)
```

```python
new_df.head()
```

```python
new_df["sex.raw"] = new_df["sex.raw"].replace(to_replace=["male","female"], value=[1,0])
new_df["mort"] = new_df["mort"].replace(to_replace=["alive","dead","unknown"], value=[1,0,-1])
new_df["smoking_status"] = new_df["smoking_status"].replace(to_replace=["former","active","never","unknown"], value=[2,1,0,-1])
new_df["stage_at_diagnosis"] = new_df["stage_at_diagnosis"].replace(to_replace=["I","II","IIIA"], value=[1,2,3])
new_df["Treatment"] = new_df["Treatment"].replace(to_replace=["Adjuvant treatment","No Adjuvant treatment"], value=[1,0])
new_df["Adjuvant_chemotherapy"] = new_df["Adjuvant_chemotherapy"].replace(to_replace=["Adjuvant_chemotherapy","No_chemotherapy"], value=[1,0])
new_df["Treatment_Radiation"] = new_df["Treatment_Radiation"].replace(to_replace=["Radiation","No_Radiation"], value=[1,0])
new_df["Recurrence_status"] = new_df["Recurrence_status"].replace(to_replace=["Recurrence","No Recurrence"], value=[1,0])
new_df.head()
```

```python
for col in colname:
    if col not in ["id","age", "days"]:
        print(new_df[col].value_counts())
```
